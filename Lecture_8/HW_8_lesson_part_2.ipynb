- Воспользуемся правилом нахождения производной от частного: 


$$\frac{d}{d x}\left(\frac{u}{v}\right)=\frac{v \frac{d u}{d x}-u \frac{d v}{d x}}{v^{2}}$$
Где $u = x$ и $ v = |x| + 1$.

Следовательно, в нашем случае имеем:

$\frac{d}{d x}\left(\frac{x}{1+|x|}\right)= \frac{(1+|x|)\left(\frac{d}{d x}(x)\right)-x\left(\frac{d}{d x}(1+|x|)\right)}{(1+|x|)^{2}}= \frac{1+|x|-x\left(\frac{d}{d x}(1+|x|)\right)}{(1+|x|)^{2}}=\frac{1+|x|-x\left(\frac{d}{d x}(|x|)+0\right)}{(1+|x|)^{2}} = $

- Запишем $\frac{d}{d x}(|x|)$ след. образом  $\frac{d}{d x}(|x|)=\frac{d|u|}{d u} \frac{d u}{d x}$, где $u=x$, $\frac{d}{d u}(|u|)=\frac{u}{|u|}$

$= \frac{1+|x|-\frac{x\left(\frac{d}{d x}(x)\right)}{|x|}}{(1+|x|)^{2}}= \frac{1+|x|-\frac{x^{2}\left(\frac{d}{d x}(x)\right)}{|x|}}{(1+|x|)^{2}}= \frac{1+|x|-1 \mid \frac{x^{2}}{|x|}}{(1+|x|)^{2}}= \frac{1}{(1+|x|)^{2}}$

$$f'(x) = \frac{1}{(1+|x|)^{2}}$$

```
import numpy as np
import matplotlib.pyplot as plt
```

1. Identity. f(x) = x

f'(x) = 1

Область значений данной функции равна её области определений - множеству всех действительных чисел.


```
def f(x):
    return x

a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Identity')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Функция Identity монотонна

2. Binary step/Единичная ступенька. $f(x)=\left\{\begin{array}{ll}0 & x<0 \\ 1 & x \geqslant 0\end{array}\right.$

Производная от данной функции будет равна нулю, так как значение функции всегда константа.

Область значений будет равна двум числам - 0 и 1

```
def f(x):
  x = x >= 0
  return x

a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Binary step')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Функция Binary step монотонна

3. **Hyperbolic tangent/Гиперболический тангенс**. $f(x)=\operatorname{th}(x)=\frac{\left(e^{x}-e^{-x}\right)}{\left(e^{x}+e^{-x}\right)}$

Саму функцию можно представить как $\frac{\left(e^{2x}-1\right)}{\left(e^{2x}+1\right)}$

Соответственно, f'(x) = $\frac{\left((e^{2x}-1)'(e^{2x}+1)-(e^{2x}-1)(e^{2x}+1)'\right)}{\left(e^{2x}+1\right)^2}=\frac{\left(4e^{2x}\right)}{\left(e^{2x}+1\right)^2}$

Область значений функции f(x) ∈ (-1,1), то есть принадлежит отрезку всех действительных чисел от -1 до 1.

```
def f(x):
  x = (np.e**x-np.e**(-x))/(np.e**x+np.e**(-x))
  return x
a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Hyperbolic tangent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

4. **ReLU**. $f(x)=\left\{\begin{array}{ll}0 & x<0 \\ x & x \geqslant 0\end{array}\right.$

$f'(x)=\left\{\begin{array}{ll}0 & x<0 \\ 1 & x \geqslant 0\end{array}\right.$

Область значений f(x) ∈ [0,∞)

```
def f(x):
  x_less_zero = x[x<0]*0
  x = np.concatenate((x_less_zero,x[x>=0]))
  return x
a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Функция ReLU монотонна


5. **Softplus**. $f(x)=\ln \left(1+e^{x}\right)$

$f'(x)=\frac{\left(\ln \left(1+e^{x}\right)\right)'}{\left(\ln \left(1+e^{x}\right)\right)}=\frac{e^{x}}{\ln \left(1+e^{x}\right)}$

Область значений f(x) ∈ (0,∞)

```
def f(x):
  x = np.log(1 + np.e**x)
  return x
a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Softplus')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

функция Softplus монотонна

6. **Синусоида**. $f(x)=\sin (x)$

 $f'(x)=\cos (x)$

Область значений f(x) ∈ [-1,1]

```
def f(x):
  x = np.sin(x)
  return x
a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Синусоида')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Функция Синусоида не монотонна
